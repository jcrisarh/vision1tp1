1. Implementar el algoritmo de pasaje a coordenadas cromáticas para librarnos de las variaciones de contraste.

```python
import cv2 as cv
import matplotlib.pyplot as plt
import os
import numpy as np

def read_original_images(image_name):
    image = cv.imread(image_name)
    return image
```

```python
def change_to_rgb(image):
    imgRGB = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    return imgRGB
```

```python
def coord_crom(image):
    img = image.copy()
    imgfloat=img.astype(np.float32)
    imgfloat2=imgfloat.copy()
    for i in range(imgfloat.shape[0]):
        for j in range(imgfloat.shape[1]):
            r,g,b = imgfloat[i,j]
            if (imgfloat[i,j] == [0,0,0]).all():
                imgfloat2[i,j] = [0,0,0]
                
            else:
                r = imgfloat[i,j][0]
                g = imgfloat[i,j][1]
                b = imgfloat[i,j][2]
                sumrgb = r + g + b
                

                rt = np.divide(r,sumrgb)
                gt = np.divide(g,sumrgb)
                bt = np.divide(b,sumrgb)
                imgfloat2[i,j] = [rt, gt, bt]
    imgcr = (255 * imgfloat2).round().astype(np.uint8) 

    return imgcr
```

```python
img1 = read_original_images("CoordCrom_1.png")
img1RGB = change_to_rgb(img1) 
img1crom = coord_crom(img1RGB)

fig, ax = plt.subplots(1,3, figsize=(20,12))
ax[0].imshow(img1)
ax[0].set_title('Imagen BGR')

ax[1].imshow(img1RGB);
ax[1].set_title('Imagen RGB')

ax[2].imshow(img1crom);
ax[2].set_title('Imagen coordenadas cromáticas')
```

```python
img2 = read_original_images("CoordCrom_2.png")
img2RGB = change_to_rgb(img2) 
img2crom = coord_crom(img2RGB)

fig, ax = plt.subplots(1,3, figsize=(20,12))
ax[0].imshow(img2)
ax[0].set_title('Imagen BGR')

ax[1].imshow(img2RGB);
ax[1].set_title('Imagen RGB')

ax[2].imshow(img2crom);
ax[2].set_title('Imagen coordenadas cromáticas')
```
