```python
import cv2 as cv
import matplotlib.pyplot as plt
import os
import numpy as np

def read_gray_scale(image_name):
    image = cv.imread(image_name, cv.IMREAD_GRAYSCALE)
    return image
    
img1 = read_gray_scale("img1_tp.png")
img2 = read_gray_scale("img2_tp.png")

fig, ax = plt.subplots(1,2, figsize=(20,12))
ax[0].imshow(img1)
ax[0].set_title('img1_tp gray scale')

ax[1].imshow(img2)
ax[1].set_title('img2_tp gray scale')

```

```python

hist1 = cv.calcHist([img1], [0], None, [256], [0, 256])

hist2 = cv.calcHist([img2], [0], None, [256], [0, 256])

fig, ax = plt.subplots(1,2, figsize=(18,7))
ax[0].plot(hist1)
ax[0].set_title('Histograma de img1_tp en escala de grises')
ax[0].set_xlabel('Valor del píxel')
ax[0].set_ylabel('Frecuencia')

ax[1].plot(hist2)
ax[1].set_title('Histograma de img2_tp en escala de grises')
ax[1].set_xlabel('Valor del píxel')
ax[1].set_ylabel('Frecuencia')
```

```python
unique1, counts1 = np.unique(img1, return_counts=True)
result1 = np.column_stack((unique1, counts1))
unique2, counts2 = np.unique(img2, return_counts=True)
result2 = np.column_stack((unique2, counts2))
```

```python
(result1==result2).all()
```

```python
(hist1==hist2).all()
```

Como se puede observar, el histograma de ambas imagenes es igual pero las imagenes son distintas. Los histogramas son simples y eficientes, ayudan a capturar información valiosa sobre la distribución de intensidades de píxeles en una imagen, ayudan en el ajuste de brillo y constraste y pueden ser utiles en el preprocesamiento y normalizacion, sin embargo, hay una perdida de informacion espacial por lo que se puede ver, en este caso no se podria utilizar como una unica feature pero podria ser usado en combinacion con otras features para una clasificacion adecuada.
